```python
df = pd.read_csv("Production, Raw Mat.csv", header=None)
print(df.shape)
print(df.to_string())  # see everything — find where table 2 starts
```

```python
# Find the blank row that separates the two tables
blank_rows = df[df.isnull().all(axis=1)].index.tolist()
print("Blank rows at:", blank_rows)  # e.g., [15] means split at row 15
```

```python
split = blank_rows[0]  # e.g., row 15

raw_prod = df.iloc[:split].copy()
raw_mat  = df.iloc[split+1:].copy()  # +1 to skip the blank row
```

```python
# First row is header
prod = raw_prod.copy()
prod.columns = prod.iloc[1]    # set header
prod = prod.iloc[2:].reset_index(drop=True)

# Separate the total row (last row)
prod_total = prod.iloc[[-1]]   # keep for reference
prod_data  = prod.iloc[:-1]    # actual data without total

prod_data.columns.name = None
prod_data.head()
```

```python
raw = raw_mat.copy()
raw.columns = raw.iloc[0]      # set header
raw = raw.iloc[1:].reset_index(drop=True)

# Last 2 rows = total + alt raw mat percentage
raw_total      = raw.iloc[[-2]]   # total row
raw_pct        = raw.iloc[[-1]]   # % alt raw mat row
raw_data       = raw.iloc[:-2]    # actual data

raw_data.columns.name = None
raw_data.head()
```

```python
def clean_columns(df):
    cols = list(df.columns)
    cols[0] = "name"
    cols[1] = "unit"
    df.columns = cols
    return df

prod_data = clean_columns(prod_data)
raw_data  = clean_columns(raw_data)
```

```python
time_cols = [c for c in prod_data.columns if c not in ["name", "unit"]]

prod_long = prod_data.melt(
    id_vars=["name", "unit"],
    value_vars=time_cols,
    var_name="period",
    value_name="value"
)

raw_long = raw_data.melt(
    id_vars=["name", "unit"],
    value_vars=time_cols,
    var_name="period",
    value_name="value"
)
```

```python
for df in [prod_long, raw_long]:
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    df["value"] = df["value"].fillna(0)
```

```python
import re

# Vietnamese character set
viet_chars = (
    r'àáâãèéêìíòóôõùúýăđơư'
    r'ạảấầẩẫậắằẳẵặẹẻẽếềểễệ'
    r'ỉịọỏốồổỗộớờởỡợụủứừửữự'
    r'ỳỵỷỹÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚÝĂĐƠƯ'
)

def remove_vietnamese(text):
    if not isinstance(text, str):
        return text

    # Remove name0, name1 prefix
    text = re.sub(r'^name\d+', '', text)

    # Pattern 1: Remove last Vietnamese parenthesis → (Sản phẩm ĐV)
    viet_paren = rf'\s*\([^)]*[{viet_chars}][^)]*\)'
    text = re.sub(viet_paren, '', text)

    # Pattern 2: Remove slash + Vietnamese → / Đá vôi (canxit)
    # Cuts everything from / onwards if what follows contains Vietnamese
    slash_viet = rf'\s*/[^/]*[{viet_chars}].*'
    text = re.sub(slash_viet, '', text)

    return text.strip()

# Test
samples = [
    "Production (DV Tile) (Sản phẩm ĐV)",
    "Production (YB Powder) (bột bán cho Yên Bình)"
]

for s in samples:
    print(remove_vietnamese(s))
# → Production (DV Tile)
# → Production (YB Powder)
```

```python
# Clean the name column
prod_long["name"] = prod_long["name"].apply(remove_vietnamese)
raw_long["name"]  = raw_long["name"].apply(remove_vietnamese)

# If column headers themselves have Vietnamese
prod_long.columns = [remove_vietnamese(c) for c in prod_long.columns]
raw_long.columns  = [remove_vietnamese(c) for c in raw_long.columns]
```
