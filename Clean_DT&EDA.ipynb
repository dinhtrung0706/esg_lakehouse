```python
raw = raw_mat.copy()
raw.columns = raw.iloc[0]      # set header
raw = raw.iloc[1:].reset_index(drop=True)

# Last 2 rows = total + alt raw mat percentage
raw_total      = raw.iloc[[-2]]   # total row
raw_pct        = raw.iloc[[-1]]   # % alt raw mat row
raw_data       = raw.iloc[:-2]    # actual data

raw_data.columns.name = None
raw_data.head()
```

```python
import re

# Vietnamese character set
viet_chars = (
    r'àáâãèéêìíòóôõùúýăđơư'
    r'ạảấầẩẫậắằẳẵặẹẻẽếềểễệ'
    r'ỉịọỏốồổỗộớờởỡợụủứừửữự'
    r'ỳỵỷỹÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚÝĂĐƠƯ'
)

def remove_vietnamese(text):
    if not isinstance(text, str):
        return text

    # Remove name0, name1 prefix
    text = re.sub(r'^name\d+', '', text)

    # Pattern 1: Remove last Vietnamese parenthesis → (Sản phẩm ĐV)
    viet_paren = rf'\s*\([^)]*[{viet_chars}][^)]*\)'
    text = re.sub(viet_paren, '', text)

    # Pattern 2: Remove slash + Vietnamese → / Đá vôi (canxit)
    # Cuts everything from / onwards if what follows contains Vietnamese
    slash_viet = rf'\s*/[^/]*[{viet_chars}].*'
    text = re.sub(slash_viet, '', text)

    return text.strip()

# Test
samples = [
    "Production (DV Tile) (Sản phẩm ĐV)",
    "Production (YB Powder) (bột bán cho Yên Bình)"
]

for s in samples:
    print(remove_vietnamese(s))
# → Production (DV Tile)
# → Production (YB Powder)
```

```python
# Clean the name column
prod_long["name"] = prod_long["name"].apply(remove_vietnamese)
raw_long["name"]  = raw_long["name"].apply(remove_vietnamese)

# If column headers themselves have Vietnamese
prod_long.columns = [remove_vietnamese(c) for c in prod_long.columns]
raw_long.columns  = [remove_vietnamese(c) for c in raw_long.columns]
```

```python
df = pd.read_csv("Water.csv", header=None)
print(df.shape)
print(df.to_string())  # see everything — find where table 2 starts
```

```python
# Row 0 is junk, row 1 is the real header
df.columns = df.iloc[1]        # set real header
df = df.iloc[2:].reset_index(drop=True)  # drop rows 0 and 1
df.columns.name = None

# Rename first two cols
df = df.rename(columns={df.columns[0]: "attribute", df.columns[1]: "unit"})
```

```python
df
```

```python
# Drop fully blank rows
df = df.dropna(how="all")

# Drop Vietnamese-only note rows (no unit, no numeric data)
note_mask = df["unit"].isna() & df.iloc[:, 2:].isna().all(axis=1)
df = df[~note_mask].reset_index(drop=True)
```

```python
df
```

```python
# Find second header row (contains "Water Recycle" in attribute col)
header2_idx = df[df["attribute"].str.contains("Water Recycle", na=False) &
                 df["unit"].str.contains("unit", na=False)].index[0]

table1 = df.iloc[:header2_idx].copy()   # Water Withdrawal
table2_raw = df.iloc[header2_idx:].copy()

# Fix table 2 headers
table2 = table2_raw.copy()
table2.columns = table2.iloc[0]
table2 = table2.iloc[1:].reset_index(drop=True)
table2 = table2.rename(columns={table2.columns[0]: "attribute", table2.columns[1]: "unit"})
```

```python
table1
```

```python
# Find the index of the first "Ko có nước ngầm" row
cut_idx = table2[table2["attribute"].str.contains("Ko c", na=False, case=False)].index[0]

# Keep only rows above it
table2 = table2.iloc[:cut_idx].reset_index(drop=True)
```

```python
table2
```

```python
viet_chars = (
    r'àáâãèéêìíòóôõùúýăđơư'
    r'ạảấầẩẫậắằẳẵặẹẻẽếềểễệ'
    r'ỉịọỏốồổỗộớờởỡợụủứừửữự'
    r'ỳỵỷỹÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚÝĂĐƠƯ'
)

def clean_attribute(text):
    if not isinstance(text, str):
        return text
    text = re.sub(r'^name\d+', '', text)
    text = re.sub(rf'\s*\([^)]*[{viet_chars}][^)]*\)', '', text)
    text = re.sub(rf'\s*/[^/]*[{viet_chars}].*', '', text)
    return text.strip()

for t in [table1, table2]:
    t["attribute"] = t["attribute"].apply(clean_attribute)
```

```python
drop_periods = ["YTD", "Q1", "Q2", "Q3", "Q4"]

def melt_and_filter(df, table_name):
    time_cols = [c for c in df.columns if c not in ["attribute", "unit"]]
    long = df.melt(id_vars=["attribute", "unit"],
                   value_vars=time_cols,
                   var_name="period", value_name="value")
    long = long[~long["period"].str.strip().str.upper().isin(drop_periods)]
    long["value"] = pd.to_numeric(long["value"], errors="coerce")
    long["table"] = table_name
    return long

withdrawal_long = melt_and_filter(table1, "Water Withdrawal")
recycle_long    = melt_and_filter(table2, "Water Recycle")
```

```python
withdrawal_long.to_csv("water_withdrawal_clean.csv", index=False)
recycle_long.to_csv("water_recycle_clean.csv", index=False)
```
